```python
def student_details():
    
    #student_name=["a","b","c","d","e","f","g","h","i","j"]
    #semester=[1,2,5,3,6,7,3,8,4,5]
    #math_marks=[20,30,35,26,18,37,40,27,13,24]
    #mechanical_marks=[20,30,35,26,18,37,40,27,13,24]
    #Electrical_marks=[20,30,35,26,18,37,40,27,13,24]
    #Python_marks=[20,30,35,26,18,37,40,27,13,24]
    #physics_marks=[20,30,35,26,18,37,40,27,13,24]
    
    no_of_students=int(input("Enter the number of students:"))
    
    student_name_ls=[]
    semester_ls=[]
    math_marks_ls=[]
    mechanical_marks_ls=[]
    Electrical_marks_ls=[]
    Python_marks_ls=[]
    physics_marks_ls=[]
    student_usn_ls=[]
    
    for i in range (no_of_students):
        student_name=input("Enter the student's name:")
        student_name_ls.append(student_name)
        
        semester=int(input("Enter the semester:"))
        semester_ls.append(semester)
        
        math_marks=int(input("Enter the math marks:")) 
        math_marks_ls.append(math_marks)
        
        mechanical_marks=int(input("Enter the mechanical marks:"))
        mechanical_marks_ls.append(mechanical_marks)
        
        Electrical_marks=int(input("Enter the Electrical marks:"))
        Electrical_marks_ls.append(Electrical_marks)
        
        Python_marks=int(input("Enter the Python marks:"))
        Python_marks_ls.append(Python_marks)
        
        physics_marks=int(input("Enter the physics marks:"))
        physics_marks_ls.append(physics_marks)
        
        student_usn=input("Enter the student usn:")
        student_usn_ls.append(student_usn)
        
    d=dict()
    for i in range(no_of_students):            
        d[student_name_ls[i]]=[semester_ls[i],student_usn_ls[i],
                    math_marks_ls[i],mechanical_marks_ls[i],
                    Electrical_marks_ls[i],Python_marks_ls[i],
                    physics_marks_ls[i]]
    print("The student details are in the following order")
    print("name: Semester usn math mechanical Electrical Python Physics")    
    print(d)    
            
    total_math_marks=0
    for j in math_marks_ls:
        total_math_marks=total_math_marks+j
    avg_math_marks=total_math_marks/no_of_students 
    print("Average marks for maths is",avg_math_marks)
        
    total_mechanical_marks=0
    for j in mechanical_marks_ls:
        total_mechanical_marks=total_mechanical_marks+j
    avg_mechanical_marks=total_mechanical_marks/no_of_students
    print("Average marks for mechanical is",avg_mechanical_marks)
        
    total_Electrical_marks=0
    for j in Electrical_marks_ls:
        total_Electrical_marks=total_Electrical_marks+j
    avg_Electrical_marks=total_Electrical_marks/no_of_students
    print("Average marks for electrical is",avg_Electrical_marks)
        
    total_Python_marks=0
    for j in Python_marks_ls:
        total_Python_marks=total_Python_marks+j
    avg_Python_marks=total_Python_marks/no_of_students
    print("Average marks for python is",avg_Python_marks)
        
    total_physics_marks=0
    for j in physics_marks_ls:
        total_physics_marks=total_physics_marks+j
    avg_physics_marks=total_physics_marks/no_of_students
    print("Average marks for physics is",avg_physics_marks)
        
    for j in range(no_of_students):
        mat=math_marks_ls[j]
        ph=physics_marks_ls[j]
        mech=mechanical_marks_ls[j]
        eee=Electrical_marks_ls[j]
        pyth=Python_marks_ls[j]
            
        if mat<avg_math_marks:
            print(student_name_ls[j],"is less than average in maths")
        if ph<avg_physics_marks:
            print(student_name_ls[j],"is less than average in physics")
        if mech<avg_mechanical_marks:
            print(student_name_ls[j],"is less than average in mechanical")
        if eee<avg_Electrical_marks:
            print(student_name_ls[j],"is less than average in electrical")
        if pyth<avg_Python_marks:
            print(student_name_ls[j],"is less than average in python")
        
    max_math=max(math_marks_ls) 
    pos_max_math=math_marks_ls.index(max_math) 
    print("The highest marks in math is",max_math,"by",student_name_ls
          [pos_max_math])
    
    max_phy=max(physics_marks_ls) 
    pos_max_phy=physics_marks_ls.index(max_phy) 
    print("The highest marks in physics is",max_phy,"by",student_name_ls
          [pos_max_phy])
    
    max_python=max(Python_marks_ls) 
    pos_max_python=Python_marks_ls.index(max_python) 
    print("The highest marks in python is",max_python,"by",student_name_ls
          [pos_max_python])
    
    max_eee=max(Electrical_marks_ls) 
    pos_max_eee=Electrical_marks_ls.index(max_eee) 
    print("The highest marks in Electrical is",max_eee,"by",student_name_ls
          [pos_max_eee])
    
    max_mech=max(mechanical_marks_ls) 
    pos_max_mech=mechanical_marks_ls.index(max_mech) 
    print("The highest marks in mechanical is",max_mech,"by",student_name_ls
          [pos_max_mech])               
                                
student_details()              
            
        
        
            
                              
    
```
